```python
from quantum_music.circuit_functions import (
    get_example_circuit,
    get_cummulative_state_vectors,
    get_circuits_by_column
)
from quantum_music.Jukebox import Jukebox
from ipywidgets import Output, HBox
from qiskit.visualization import plot_state_qsphere
```

```python
circuit = get_example_circuit()
```

```python
jukebox = Jukebox(circuit)
```

```python
sub_circuits = get_circuits_by_column(circuit)
state_vectors = get_cummulative_state_vectors(sub_circuits)

out1 = Output(layout={'flex-direction': 'row'})
with out1:
    display(circuit.draw())
    display(sub_circuits[0].draw())
out2 = Output(layout={'flex-direction': 'row'})
with out2:
    display(plot_state_qsphere(state_vectors[0]))
HBox([out1, out2])
```
