```python
from quantum_music.circuit_functions import (
    get_example_circuit,
    get_circuits_by_column,
    get_cummulative_state_vectors,
    play_notes,
    play_notes_from_state_vector,
    get_note,
    get_notes
)
from numpy import pi
```

```python
circuit = get_example_circuit()
state_vectors = get_cummulative_state_vectors(get_circuits_by_column(circuit))
assert len(state_vectors) > 0

state_vector = state_vectors[1]
play_notes_from_state_vector(state_vector, show_vectors=True)
notes = get_notes(state_vector)
print('notes', notes)
```

```python
play_notes_from_state_vector(state_vector, use_volume=False)
```

```python
# Test get_note()
phases_c_scale = [
    (0, 'C4'),
    (pi/4, 'D4'),
    (pi/2, 'E4'),
    (-pi/4, 'C5'),
    (-pi/2, 'B4')
]
for (phase, note_name) in phases_c_scale:
    note = get_note(phase)
    print(f'Got {note[0]}, expected {note_name}')
    print('Correct? ', note[0] == note_name)
```
